```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# RGB image
image = cv2.imread("flowers.jpg")
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

print("shape : ", rgb_image.shape)
plt.xticks([])
plt.yticks([])
plt.imshow(rgb_image)
plt.show()
```

```python
# GrayScaling image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image = np.expand_dims(gray_image, axis=2)

print("shape : ", gray_image.shape)
plt.xticks([])
plt.yticks([])
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
r_image = rgb_image[:, :, 0]
r_image = np.expand_dims(r_image, axis=2)

print("shape : ", r_image.shape)
plt.xticks([])
plt.yticks([])
plt.imshow(r_image, cmap='Reds')
plt.show()
```

```python
g_image = rgb_image[:, :, 1]
g_image = np.expand_dims(g_image, axis=2)

print("shape : ", g_image.shape)
plt.xticks([])
plt.yticks([])
plt.imshow(g_image, cmap='Greens')
plt.show()
```

```python
b_image = rgb_image[:, :, 2]
b_image = np.expand_dims(b_image, axis=2)

print("shape : ", b_image.shape)
plt.xticks([])
plt.yticks([])
plt.imshow(b_image, cmap='Blues')
plt.show()
```

```python
# Sobel-filtering Image
sobel_image = cv2.Sobel(src=gray_image,ddepth=-1, dx=1,dy=0,ksize=3) 
sobel_image = np.expand_dims(sobel_image, axis=2)

print("shape : ", sobel_image.shape)
plt.xticks([])
plt.yticks([])
plt.imshow(sobel_image, cmap='gray')
plt.show()
```

```python
# RGB-Gray-Sobel concate
concat_images = np.concatenate([rgb_image, gray_image, sobel_image], axis=2)
concat_images.shape
```
